##  Checking distribution of the variables

```python
l = data.columns.values
number_of_columns=len(l)
number_of_rows = len(l)/3
plt.figure(figsize=(number_of_columns , 3*number_of_rows))
for i in range(0,len(l)):
    plt.subplot(number_of_rows , number_of_columns/3, i+1)
    sns.set_style('whitegrid')
    sns.distplot(data[l[i]], color='green', kde = True)
    ax = sns.distplot(data[l[i]], color='green',  kde = True)
    ax.set_ylabel(data.columns[i])
    plt.tight_layout()
```

* 'pH' and 'density' appears to be normally distributed

## To check Correlation

```python
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(), annot=True, cmap="Greens")
plt.show()
```

* Here we can infer that "fixed acidity" has strong positive correlation with "citric acid" and "density".
* "volitile acidity" and "sulphates" has almost no correlation with "residual sugar".
*  "free sulphur dioxide" and "chlorides" has almost no correlation.
* Since correlation is zero we can infer there is no linear relationship between these two predictors.

## To check Outliers

```python
l = data.columns.values
number_of_columns=len(l)
number_of_rows = len(l)/3
plt.figure(figsize=(number_of_columns , 3*number_of_rows))
for i in range(0,len(l)):
    plt.subplot(number_of_rows , number_of_columns/3, i+1)
    sns.set_style('whitegrid')
    sns.boxplot(data=data[l[i]], color='green', orient='v')
    ax = sns.boxplot(data=data[l[i]], color='green', orient='v')
    ax.set_xlabel(data.columns[i])
    plt.tight_layout()
```

* All features shows outliers

```python
for i in data.columns:
    if i =="quality":
        break
    sns.boxplot("quality",i,data=data,color='green')
    plt.show()
```

* Quality 3 and 8 are having less outliers and quality 5 and 6 has more outliers
